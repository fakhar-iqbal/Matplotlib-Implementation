```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as mp

image1 = Image.open("flg.jpeg")
#img1_array  = np.array(image1)


image2 = Image.open("pak.jpg")
#img2_array = np.array(image2)

desired_shape = (600,400)
resize_img1 = image1.resize(desired_shape)
resize_img2 = image2.resize(desired_shape)

combined  = np.concatenate((resize_img1,resize_img2),axis=1)

mp.imshow(combined)
```
